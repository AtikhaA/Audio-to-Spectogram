```python
import os
import IPython.display as ipd
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
ipd.Audio('/content/drive/MyDrive/audio/test7.wav')
```

```python
filename3 = '/content/drive/MyDrive/audio/test7.wav'
plt.figure(figsize=(15,4))
data2,sample_rate2 = librosa.load(filename3, sr=22050, mono=True, offset=0.0, duration=50, res_type='kaiser_best')
librosa.display.waveplot(data2,sr=sample_rate2, max_points=50000.0, x_axis='time', offset=0.0, max_sr=1000)
```

```python
audio_path1 = "/content/drive/MyDrive/audio/"
audio1 = os.listdir(audio_path1)
```

```python
x, sr = librosa.load(audio_path1+audio1[5], sr=44100)
 
print(type(x), type(sr))
print(x.shape, sr)
```

```python
plt.figure(figsize=(14, 5))
librosa.display.waveplot(x, sr=sr)
```

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```
